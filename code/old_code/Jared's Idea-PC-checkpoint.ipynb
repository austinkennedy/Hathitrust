## Load data
Here we load the volume-topic matrix and the words by topic

```python
import os
import pandas as pd
volume_topics = '/Users/RustyOllie/Google Drive/HTRC/Genre Analysis/20191007_topics.txt'

topic_words = '/Users/RustyOllie/Google Drive/HTRC/Genre Analysis/20191007_keys.txt'
topics = pd.read_csv(topic_words, sep = '\t', lineterminator = '\n', header=None)
# topics.drop(columns = 0, inplace = True)
topics.columns = ['topic_num','weight','words'] 
topics['topic_num'] = topics['topic_num'] + 1

data = pd.read_csv(volume_topics, sep = '\t', lineterminator = '\n', header=None)
data.drop(columns = 0, inplace = True)
data[1] = [string[string.rfind('/UK_data/')+9:-4] for string in data[1]]
data.columns = ['HTID'] + [i for i in range(1,61)]
print("Dimensions: " + str(data.shape))
htids = data['HTID']
data = data.drop(columns=['HTID'])
data.head()
```

## Calculate overall breadth 
Here we calculate per-topic breadth by summing the proportions of each topic across all books

```python
# pd.set_option('display.max_colwidth', -1)

bc = pd.DataFrame(data.columns[0:61], columns = ['Topic'])

u = []
for topic in range(1,61):
    u.append(sum(data[topic]))
    
a = pd.concat([bc, pd.DataFrame(u)],axis=1)
a.rename(columns={0: "breadth"},inplace=True)
a = pd.merge(a,topics,left_on='Topic',right_on='topic_num')
a = a.drop(columns=['weight', 'topic_num'])
a.sort_values('breadth', ascending=False)
```

### Next we round-down each volume year to the nearest quarter centutry (1500, 1525, 1550....)

```python
import pickle

volume_topics = '/Users/RustyOllie/Google Drive/HTRC/Genre Analysis/20191007_topics.txt'

data = pd.read_csv(volume_topics, sep = '\t', lineterminator = '\n', header=None)
data.drop(columns = 0, inplace = True)
data[1] = [string[string.rfind('/UK_data/')+9:-4] for string in data[1]]
data.columns = ['HTID'] + [i for i in range(1,61)]

metapath = '/Users/RustyOllie/Google Drive/HTRC/Genre Analysis/metadata.p'
metadata = pickle.load(open(metapath, 'rb'))

metadata['Year_rounded'] = pd.to_numeric(metadata['Year'])
metadata['Year'] = pd.to_numeric(metadata['Year'], downcast='signed')

metadata.loc[(metadata['Year'] >= 1500) & (metadata['Year'] < 1525), 'Year_rounded'] = 1500
metadata.loc[(metadata['Year'] >= 1525) & (metadata['Year'] < 1550), 'Year_rounded'] = 1525
metadata.loc[(metadata['Year'] >= 1550) & (metadata['Year'] < 1575), 'Year_rounded'] = 1550
metadata.loc[(metadata['Year'] >= 1575) & (metadata['Year'] < 1600), 'Year_rounded'] = 1575

metadata.loc[(metadata['Year'] >= 1600) & (metadata['Year'] < 1625), 'Year_rounded'] = 1600
metadata.loc[(metadata['Year'] >= 1625) & (metadata['Year'] < 1650), 'Year_rounded'] = 1625
metadata.loc[(metadata['Year'] >= 1650) & (metadata['Year'] < 1675), 'Year_rounded'] = 1650
metadata.loc[(metadata['Year'] >= 1675) & (metadata['Year'] < 1700), 'Year_rounded'] = 1675

metadata.loc[(metadata['Year'] >= 1700) & (metadata['Year'] < 1725), 'Year_rounded'] = 1700
metadata.loc[(metadata['Year'] >= 1725) & (metadata['Year'] < 1750), 'Year_rounded'] = 1725
metadata.loc[(metadata['Year'] >= 1750) & (metadata['Year'] < 1775), 'Year_rounded'] = 1750
metadata.loc[(metadata['Year'] >= 1775) & (metadata['Year'] < 1800), 'Year_rounded'] = 1775

metadata.loc[(metadata['Year'] >= 1800) & (metadata['Year'] < 1825), 'Year_rounded'] = 1800
metadata.loc[(metadata['Year'] >= 1825) & (metadata['Year'] < 1850), 'Year_rounded'] = 1825
metadata.loc[(metadata['Year'] >= 1850) & (metadata['Year'] < 1875), 'Year_rounded'] = 1850
metadata.loc[(metadata['Year'] >= 1875) & (metadata['Year'] < 1900), 'Year_rounded'] = 1875

def fix_htid(row):
    return row['HTID'].replace(":","+").replace("/", "=")

metadata['HTID'] = metadata.apply(fix_htid, axis=1)
df2 = pd.merge(data, metadata, on='HTID', how='inner').drop(columns = ['oclc', 'Year'])
```

### Here I get the breadth by quarter century periods, I save them into a dictionary *breadth_by_year*

### Next I create a dataframe *b* that has the raw breadth values by time period

```python
years = [1500, 1525, 1550, 1575, 1600, 1625, 1650, 1675, 1700, 1725, 1750, 1775, 1800, 1825, 1850, 1875]
breadth_by_year = {}
for year in years:
    d = df2[df2['Year_rounded'] == year].drop(columns=['HTID','Year_rounded'])

    bc = pd.DataFrame(data.columns[1:61], columns = ['Topic'])
    u = []
    for topic in range(1,61):
        u.append(sum(d[topic]))
    a = pd.concat([bc, pd.DataFrame(u)],axis=1)
    a.rename(columns={0: "breadth"},inplace=True)
    breadth_by_year[year] = a
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

u = []
for year in years:
    u.append(breadth_by_year[year]['breadth'])
b = pd.DataFrame(u)
b = b.transpose()
b.columns = years
b.head()
```
