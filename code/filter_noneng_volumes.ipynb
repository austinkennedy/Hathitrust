```python
print('Loading Packages')
import pandas as pd
import numpy as np
import config


print('Loading Data')

data = pd.read_csv('../temporary/topic_weights_' + config.topic_weights + '.csv')
title_data = pd.read_csv('../input/volume_titles.csv', encoding='latin-1')
titles = title_data[['HTID', '245a', '245b']].rename(columns={'245a': 'title_1',
                                                              '245b': 'title_2'})


data.head()
```

```python
topics = [18,32,49]
```

```python
def top_volumes_cat(data, titles, n, topic):
    """
    Select the top `n` volumes for topic 'topic', merges with title data

    Parameters:
    data (pandas dataframe): the cleaned Hathitrust data
    n (int): the number of volumes you want to return
    topic (int): the category number

    Returns: A dataframe of HTIDS and the category score
    """

    sorted = data.sort_values(by = f'{topic}', ascending=False)
    selected = sorted[['HTID', f'{topic}']].head(n)
    selected = selected.merge(titles, on = 'HTID', how = 'left')
    return selected
```

```python
dfs = {}
n = 3000
for t in topics:
    dfs[t] = top_volumes_cat(data, titles, topic = t, n = n)
```

```python
#export data
for t in topics:
    dfs[t].to_csv('../output/random/top_' + str(n) + '_volumes_topic_' + str(t) + '.csv')
```
