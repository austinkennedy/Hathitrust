```python
#Import Packages
print('Importing Packages')
import pandas as pd
import pickle
```

```python
volumes = pd.read_csv('../temporary/volumes_scores.csv')
volumes
```

```python
metapath = '../input/metadata.p'
metadata = pickle.load(open(metapath, 'rb'))

metadata['Year_rounded'] = pd.to_numeric(metadata['Year'])
metadata['Year'] = pd.to_numeric(metadata['Year'], downcast='signed')
def fix_htid(row):
    return row['HTID'].replace(":","+").replace("/", "=")

metadata['HTID'] = metadata.apply(fix_htid, axis=1)
```

```python
h_progress = volumes[(volumes['Year'] >= 1700) & (volumes['Year'] <= 1850)]
h_progress = h_progress[(h_progress['progress_percentile'] >= 0.99) & (h_progress['industry_3_percentile'] >= 0.99)]
h_progress = h_progress.sort_values(by=['progress_percentile', 'industry_3_percentile'], ascending=False)
h_progress
```

```python
example_volumes = h_progress[['HTID', 'Year', 'industry_3_percentile', 'progress_percentile']]
example_volumes.to_csv('../output/example_volumes.csv', index=False)
```
