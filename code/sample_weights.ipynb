```python
import pandas as pd
```

```python
sample_dict = {}
samples = ['sample_1', 'sample_2', 'sample_3', 'sample_4']
for sample in samples:
    sample_dict[sample] = pd.read_csv('../temporary/volumes_' + sample + '.csv')
    sample_dict[sample].rename(columns = {'HTID_' + sample: 'HTID'}, inplace = True)
    sample_dict[sample].set_index('HTID', inplace = True)
```

```python
volumes = pd.concat(sample_dict, axis= 1)
volumes.columns = volumes.columns.droplevel()
```

```python
volumes
```

```python
volumes['Religion'] = volumes[volumes.columns[volumes.columns.str.contains('Religion')]].mean(axis=1)
volumes['Science'] = volumes[volumes.columns[volumes.columns.str.contains('Science')]].mean(axis=1)
volumes['Political Economy'] = volumes[volumes.columns[volumes.columns.str.contains('Political Economy')]].mean(axis=1)
```

```python
volumes
```

```python
volumes_averaged = volumes[['Religion', 'Science', 'Political Economy']].reset_index()
```

```python
volumes_averaged
```

```python
volumes_averaged.to_csv('../temporary/volumes.csv')
```
